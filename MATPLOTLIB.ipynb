```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,8])
ypoints=np.array([3,10])
plt.plot(xpoints,ypoints,"o")
plt.show()
```

```python
# Multiple lines with all properties i.e
#line width,line color,linestyle,marker
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1,lw='8',c='y',ls=':',marker='*')
plt.plot(y2,lw='5',c='r',ls='--',marker='o')
plt.show()
```

```python
#Labels and titles
#Fonts and Position of title
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("SPORTS WATCH DATA", loc='center')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#Subplot with title and super title
```

```python
#plot1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.title("SALES")
plt.plot(x,y)

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.title("INCOME")
plt.plot(x,y)
plt.suptitle("TRIAL BALANCE")
plt.show()
```

```python
#Scatter plots with color,size an alpha
# replace color with size
```

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()

```

```python
#Matplotlib bars with bar color,bar width, bar height
#alternate btwn height and width
```

```python
x=np.array(['A,','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x, color='hotpink',height=5)
plt.show()
```

```python
#Histogram
#can use uniform or random
```

```python
x=np.random.uniform(170,10,250)
plt.hist(x)
plt.show()
```

```python
#Pie Charts with labels,shadow,exlode,legend,with header and color
```

```python
x=np.array([35,25,40,15])
mylabels=["Music","Movies","Films","Songs"]
myexplode=[0.5,0,0.2,0]
plt.pie(x,labels=mylabels,startangle=180,explode=myexplode,shadow=True)
plt.legend(title="Pass Time:")
plt.show()
```
